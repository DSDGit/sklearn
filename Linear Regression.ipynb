```python
from sklearn.datasets import load_diabetes
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
```

### Sample Data 
Here we are creating a sample data ourselves. Here we have numbers from 0 to 30 on x-axis and their squares on y-axis

```python

x = [i for i in range(30)]

y = [i*2 for i in x]
```

```python
import numpy as np
x = np.array(x)[:, np.newaxis]
```

### Plotting 
Plotting the data with matplotlib to show the data points

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x, y, color='green')
plt.xlabel('Numbers')
plt.ylabel('Squarred')
plt.show()
```

### Linear Regression
Model fitting

```python
#Creating a LinearRegression model and fitting it to our sample data
model = LinearRegression()
model.fit(x,y)
```

```python
#Model Weights
model.coef_
```

```python
predicted = model.predict(x)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x, y, color='green')
plt.plot(x, predicted, color='magenta', linewidth=3)

plt.xlabel('Numbers')
plt.ylabel('Squarred')
plt.show()
```

### 1. Reading Data 

```python
dataset = load_diabetes()
print dataset.DESCR
```

```python
X,y = dataset.data, dataset.target
```

```python
X.shape
```

```python
y.shape
```

```python
#Considering only 2nd Feature
X = X[:, np.newaxis, 2]
```

### 2. Visualize Data

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,8))
plt.xlabel('Body mass Index')
plt.ylabel('Diabetes score')
plt.scatter(X, y)
plt.show()
```

### 3. Linear Regression
3.1 Data split

```python
x_train, x_test, y_train, y_test = train_test_split(X,y, test_size=0.2, random_state=32)
```

```python
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
predicted = model.predict(x_test)
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_test, y_test)
plt.plot(x_test, predicted, color='green', linewidth=3)

plt.xlabel('Body mass Index')
plt.ylabel('Diabetes score')
plt.show()
```
